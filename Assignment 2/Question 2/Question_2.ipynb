```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('/content/Salary_dataset.csv')
```

```python
df
```

```python
df.columns
```

```python
X = df[['YearsExperience']]
```

```python
X
```

```python
y = df['Salary']
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = None)
```

Before Min Max Scaling and Normalization


```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
mse
```

```python
r2
```

Doing Min Max Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# Initialize scaler
scaler = MinMaxScaler()

# Fit on training data only
X_train_M = scaler.fit_transform(X_train)

# Transform test data
X_test_M = scaler.transform(X_test)
```

```python
model2 = LinearRegression()
```

```python
model2.fit(X_train_M, y_train)
```

```python
y_pred_M = model2.predict(X_test_M)
```

```python
mse_M = mean_squared_error(y_test, y_pred_M)
r2_M = r2_score(y_test, y_pred_M)
```

```python
mse_M
```

```python
r2_M
```

Doing Normalization

```python
from sklearn.preprocessing import StandardScaler
```

```python
model3 = LinearRegression()
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
model3.fit(X_train_std, y_train)
y_pred = model3.predict(X_test_std)
```

```python
mse_standard = mean_squared_error(y_test, y_pred)
r2_standard = r2_score(y_test, y_pred)
```

```python
mse_standard
```

```python
r2_standard
```
